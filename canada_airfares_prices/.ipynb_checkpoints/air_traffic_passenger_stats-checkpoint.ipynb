```python
import pandas as pd

import json

import requests
```

```python
file = "Resources/Air_Traffic_Passenger_Statistics.csv"
```

```python
air_traffic = pd.read_csv(file)
```

```python
air_traffic.head(50)
```

```python
usa_flights = air_traffic.loc[(air_traffic["GEO Region"] == "US"), :]
```

```python
usa_flights["GEO Summary"].value_counts()
```

```python
canada_flights = air_traffic.loc[(air_traffic["GEO Region"] == "Canada"), :]
```

```python
canada_flights.head()
```

```python
canada_flights["GEO Summary"].value_counts()
```

```python
canada_cleaned = canada_flights[["Year","Month","Published Airline IATA Code","GEO Summary","Activity Type Code","Price Category Code","Adjusted Passenger Count"]]
```

```python
canada_cleaned
```

```python
canada_2011_2016 = canada_cleaned.loc[(canada_cleaned["Year"] > 2010), :]
```

```python
canada_2011_2016["GEO Summary"].value_counts()
```

```python
canada_2011_2016.head(50)
```

```python
grouped = canada_2011_2016.groupby(["Year","Month"]).sum()
```

```python
dataframe = grouped.reset_index()
```

```python
dataframe
```

```python
avg_traffic = dataframe.groupby("Month").mean()
```

```python
avg_traffic = avg_traffic["Adjusted Passenger Count"]
```

```python
avg_traffic = pd.DataFrame(avg_traffic)

avg_traffic
```

```python
avg_traffic = avg_traffic["Adjusted Passenger Count"].astype(int)
```

```python
avg_traffic = pd.DataFrame(avg_traffic)
```

```python
avg_traffic
```

```python
avg_traffic = avg_traffic.reset_index()
```

```python
avg_traffic = pd.DataFrame(avg_traffic)

avg_traffic
```
